## Improving the Visibility of the Charts

* It is very important that the charts we make look beautiful to our eyes.
* The Charts should be Informative and Impactful, but it seeks attention only when it looks attractive.
* It is very important to mention the titles, labels, legends wherever it is required.
* There should be appropriate use of colors and color palettes so that charts have more story telling values.
* There should be proper shaping and sizing of the text fonts and Charts.

```python
# lets import the basic libraries
import numpy as np
import pandas as pd

# for data visualizations
import matplotlib.pyplot as plt
```

```python
# lets import the dataset
data = pd.read_csv('Datasets/employee.csv')

# lets also check the shape of the dataset
data.shape
```

```python
# lets check the head of the dataset

data.head()
```

```python
# lets check out the available themes available to apply on the charts

plt.style.available
```

### Pie Charts

```python
# Beautifying the plots
# pie chart 

size = data['Attrition'].value_counts()
labels = ['Yes', 'No']
colors = ['orange', 'yellow']
explode = [0, 0.1]
, 
plt.rcParams['figure.figsize'] = (7, 7)
plt.style.use('dark_background')
plt.pie(size, colors = colors, labels = labels, shadow = True, explode = explode, autopct = '%.2f%%')
plt.title('Attrition Analysis', fontsize = 20)
plt.axis('off')
plt.legend()
plt.show()
```

### Bar Charts

```python

plt.rcParams['figure.figsize'] = (15, 6)
plt.style.use('ggplot')

data['Age'].value_counts().plot(kind = 'bar', color = 'purple')
plt.title('Distribution of Age', fontsize = 20)
plt.xlabel('Different Categories of Age', fontsize = 20)
plt.ylabel('Count')
plt.show()
```

### Line Charts

```python
help(plt.plot)
```

```python
plt.rcParams['figure.figsize'] = (15, 6)
plt.style.use('seaborn-talk')

data['EducationField'].value_counts().plot(kind = 'line',
                                           linewidth = 2,
                                           marker = 'o',
                                           markersize = 20,
                                           markerfacecolor = 'yellow',
                                           markeredgecolor = 'green',
                                           color = 'purple')
plt.title('Distribution of Age', fontsize = 20)
plt.xlabel('Different Categories of Age', fontsize = 20)
plt.ylabel('Count')
plt.show()
```
