## Bar Plot

### Univariate Plots

* Where we plot the values of only one variable using a bar plot
* Here, We will take a Variable and we will plot the categories against their values.
* Univarite Analysis is a process of analyzing an attribute present in the data using plots.


````python
# Lets try to analyze the values present in Attrition column of the employee dataset

# here we are using value_counts() which return the values related to each of the categories present in the attribute
x = data['Attrition'].value_counts()

# printing the values
print(x)
````

* The Value Counts Function returns the values for all the categories present in that attribute

````python
# lets use the the values in x to plot 

plt.rcParams['figure.figsize'] = (16, 4)
x.plot(kind = 'bar', color ='green')

plt.title('Univariate Analysis on Attrition Column of Employee Dataset', fontsize = 12)
plt.xlabel('Attrition')
plt.ylabel('Values')
plt.show()
````

* Using the plot function we plotted these values using a Bar Chart
* It is Clearly Visible from the Bar Chart, that Most of the Employees did not attrition out of the Organization
* It is Evident now, that using Charts it becomes easy to understand data.

````python
# analyzing the attribute Department

data['Department'].value_counts().plot(kind = 'bar', color = 'lightblue')
plt.ylabel('Values')
plt.xlabel('Different Departments')
plt.xticks(rotation = 360)
plt.title('No. of Employees in each Departments', fontsize =12)
plt.show()
````

* The above Bar Plot represents the number of employees present in each of the department
* From the Above Graph, It is Clear that most of the Employees are working for Research and Development and only few employees are working in the HR Department.
* There is a Huge and visible gap in the number of employees working in different departments of an organization

````python
# analyzing the Education Field

plt.rcParams['figure.figsize'] = (16, 4)
data['EducationField'].value_counts().plot(kind = 'bar', color = 'orange')
plt.ylabel('Values')
plt.xlabel('Different Education Fields')
plt.xticks(rotation = 360)
plt.title('No. of Employees from different Education Fields', fontsize =12)
plt.show()
````

* The Above Chart represents one more example for Univariate Analysis
* The Above Chart, Represents the number of employees in different education fields
* It is very clear from the graph that most of the Employees are form Life Sciences and Medical Fields.

## Bivariate Plots

* Plotting Two Variables in one Plot.
* Bivariate Analysis leaves us with more options to visualize the patterns in data.
* We can compare two different Variables where the First Variable must be categorical in nature and the Second Variable should be continuous variable.
* These Graphs are the most commonly used Bar Graphs.
* In this Section, we will be using the Bar Function to plot the Bar plots

````python
# lets check the documentation of Bar Plots in Matplotlib
help(plt.bar)
````

````python
# lets plot Education vs Monthly Rate

plt.rcParams['figure.figsize'] = (16, 4)
plt.bar(data['EducationField'], data['MonthlyRate'], width = 0.7, align = 'center')
plt.title('Average Salary of Employees from Various Departments')
plt.show()
````

* The Above Graph is the first example for Bivariate Analysis, where we are Comparing Different Education Fields of Employees4 in an Organization with their Average Salaries.
* The Graph shows that all the Education Fields are having almost similar average salaries.
* The HR Department is having least average salary for employees in an organization.

````python
# lets plot Departments vs Monthly Rate

plt.rcParams['figure.figsize'] = (16, 4)
plt.bar(data['Department'], data['MonthlyRate'], color = 'green', width = 0.55, align = 'center')
plt.title('Average Salary of Employees from Various Education Fields')
plt.show()
````

* Here, In the Second Example for Bivariate Analysis using Bar Charts using Bar Function from matplot
* Here, The First Variable is the Department
* The Second Variable is the Average Salary of Employees in an Organization
* We can clearly see that The Human Resource Employee get lesser pay in comparison to Sales and Research and Development Employees.

## Multivariate Plots

* There are two different types of plots available to visualize Multivariate data using bar Plots
* 1. Grouped Bar Plots
* 2. Stacked Bar Plots

**We can Plot ```Multi Variate Charts``` for any data using ```Pivot_table()``` function**

#### Grouped Bar Plots

````python
# lets plot Departments vs Monthly Rate

plt.rcParams['figure.figsize'] = (16, 4)
x = data.pivot_table(index = ['Department','EducationField'], values = ['MonthlyRate'], aggfunc = 'mean')
x.plot(kind = 'bar', stacked = False)
plt.xticks(rotation = 90)
plt.show()
````

* Using the Pivot Table Function, The Data is not in good shape for best visualization.
* We can use ```Crosstab() Function``` for better visualization.

#### We can Plot ```Multi Variate Charts``` for any data using ```crosstab()``` function**

* We can visualize the data in a very organized manner which have very high story telling value

* We have two choices, we can plot them as ```Grouped Charts``` and ```Stacked Charts```

````python
# using crosstab funtion, making a cross tabular using multiple variables.
# we have used mean function to aggregate the values of Monthly rate.
# for x axis -> index: Department
# for y axis -> index: Education Field
# values (x X y) -> Monthly Rate, aggreated using mean function

x = pd.crosstab(index = data['Department'],
                columns = data['EducationField'],
                values = data['MonthlyRate'],
                aggfunc = 'mean')
x
````

##### Here, we can see the result which we have received after Using the ```Crosstab function```
* Where the ```index``` is ```Department``` attribute.
* Where the ```column``` is ```Education Field``` attribute.
* Where the ```value``` is ```Monthly Rate``` attribute where aggregation is done using ```Mean Function```. 

#### Stacked Bar Plot

````python
# lets plot a Stacked Bar Chart
# we will use x, which we have created using crosstab function

# setting the size of the plot
plt.rcParams['figure.figsize'] = (16, 4)

# keeping the Stacked option to **True**
x.plot(kind = 'bar', stacked = True)
plt.ylabel('Salary')
plt.xticks(rotation = 20)
plt.show()
````

* We can see the Above Stacked Bar plot.
* We can that Education Fields are Stacked for all the different Departments.
* The Sizes are the average Salaries of the employees.
* It is very clear to drive insights from the above graph.
* The Employees who took HR in Education are only working in HR Department.
* The Employees having Technical Degrees are not working in the HR Department

#### Grouped Bar Plot


````python
# Lets make a Grouped Bar Plot 
# we will be using the variable x which we have created using the cross tab function

# setting the size of the plot
plt.rcParams['figure.figsize'] = (16, 4)

# keeping the stacked option to **false**
x.plot(kind = 'bar', stacked = False)
plt.xticks(rotation = 20)
plt.show()
````
